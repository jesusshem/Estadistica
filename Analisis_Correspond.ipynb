```python
from __future__ import division
import pandas as pd
import numpy as np
```

```python
def df_nuevo(num_columnas, num_registros):
    columnas = []
    for _ in range(1, num_columnas + 1):
        columnas.append('x' + str(_))

    valores = []
    for _ in range(num_columnas):
        valores.append(np.random.poisson(11, size=num_registros))

    tabla = dict(zip(columnas, valores))
    
    return pd.DataFrame(tabla)
```

```python
def totales(df):
    total_filas = []
    for _ in range(df.shape[0]):
        total_filas.append(df.loc[_].sum())
        df['total'] = pd.DataFrame(total_filas)

    total_cols = []
    for _ in df.columns.tolist():
        total_cols.append(df[_].sum())
    df.loc['total'] = total_cols
    
    return df
```

```python
df = df_nuevo(5, 15)

df
```

```python
df = totales(df)

df
```

```python
cols = []
total = df['total'][-1]
for i in range(df.shape[1] - 1):
    sub_lista = []
    for j in range(df.shape[0] - 1):
        sub_lista.append(df['total'][j]*df.loc['total'][i]/total)
    cols.append(sub_lista)
```

```python
nombres_col = ['x' + str(i) for i in range(1, 6)]

df_2 = pd.DataFrame(dict(zip(nombres_col, cols)))
```
